# Chord Relative Experiments
Play with gap between chords and repeats to choose defaults.

```python
from eartrainer.drills.chord_relative import TwoChordRelativeDrill
from eartrainer.drills.base_drill import DrillContext
from eartrainer.audio.playback import make_synth_from_config
from eartrainer.config.config import load_config, validate_config

cfg = validate_config(load_config(None))
synth = make_synth_from_config(cfg)
synth.program_piano()
```

```python
# Helper to try different settings
def play_pair(key='C', scale_type='major', degrees=('2','3','4','5','6','7'), gap_ms=350, repeat_each=2, questions=4):
    ctx = DrillContext(
        key_root=key,
        scale_type=scale_type,
        reference_mode='cadence',
        degrees_in_scope=list(degrees),
        include_chromatic=False,
        min_midi=None, max_midi=None, test_note_delay_ms=300,
    )
    drill = TwoChordRelativeDrill(ctx, synth)
    drill.configure(orientation='both', inter_chord_gap_ms=int(gap_ms), repeat_each=int(repeat_each))
    # Run a few questions without UI: just call next_question() repeatedly
    for _ in range(int(questions)):
        drill.next_question()

# Example: try a slower tempo with repeats
play_pair(gap_ms=400, repeat_each=2)
```
